```python
# package and data importation

import numpy as np
import networkx as nx
import scipy.io
import matplotlib as plt
#import pcalg
from scipy.stats import norm
import math
import pingouin as pg
import pandas as pd

from sklearn.linear_model import LinearRegression
import itertools
```

```python
#data loading

data = scipy.io.loadmat('Data.mat')['D']
col = []
for i in range(data.shape[1]):
    col.append(str(i))

df = pd.DataFrame(data, columns=col)
```

```python
# coefficient

alpha = 0.01
n = data.shape[0]

def test_function(alpha,n,S):
    return norm.isf(1-alpha/2) / np.sqrt(n-np.abs(S)-3)
```

```python
# we create the initial fully connected graph

G = nx.complete_graph(range(data.shape[1]))
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
len(G.edges())
```

```python
# PC algorithm 

K = 8 # initial assumption, question to ask
nodes = np.arange(data.shape[1])


for k in range(0,K):                              # iteration over the degree k
    for n in range(data.shape[1]):                # iteration though all nodes
        for m in range(data.shape[1]):            # iteration through all remaining nodes to make pariwise
            if n != m :
                iterator = np.delete(nodes,[n,m])
                for subset in itertools.combinations(iterator,k): # we generate all possible subset
                    
                    s = []
                    for i in subset:       # loop in order to have strings to be argument for dataframe
                        s.append(str(i))
                
                    r = pg.partial_corr(data=df, x=col[n], y=col[m],
                                        covar=s,method='spearman').round(3)['r']['spearman'] # conditionnal proba
                    test = test_function(alpha,data.shape[0], k) #threshold
                    
                    if np.abs(r) <= np.abs(test) and [n,m] in G.edges():
                        print([n,m])
                        print(r)
                        G.remove_edge(n, m)
                        
                        
```
